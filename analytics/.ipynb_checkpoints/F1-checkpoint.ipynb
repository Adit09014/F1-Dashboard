```python
schedule = f1.get_event_schedule(2026)
```

```python
schedule.head()
```

```python
today = pd.Timestamp.now(tz='UTC')
print(today)
```

```python
completed_races = schedule[
    schedule['Session5Date'] < today
]
```

```python
last_3_races=completed_races.tail(3)
```

```python
from datetime import datetime
all_results = []
year = datetime.now().year

for _, race in last_3_races.iterrows():

    session = f1.get_session(
        year,
        race['EventName'],
        'R'
    )

    session.load()

    results = session.results[
        ['Abbreviation', 'Points','TeamId']
    ].copy()

    all_results.append(results)
print(all_results)
```

```python
combined_df = pd.concat(all_results)
print(combined_df)
```

```python
combined_df = pd.concat(all_results)

# Average points of last 3 races
avg_points_df = (
    combined_df
    .groupby('Abbreviation', as_index=False)
    .agg({
        'Points':'mean',
        'TeamId':'first'
        }
    )
)

# Rename columns
avg_points_df.columns = ['driver', 'points','TeamId']

# Sort by avg points
avg_points_df = avg_points_df.sort_values(
    by='points',
    ascending=False
)

```

```python
combined_df = combined_df.reset_index()
combined_df.rename(columns={"index": "DriverNumber"}, inplace=True)
```

```python
combined_df.head(10)
```

```python
avg_points_df.head(10)
```

```python
url="https://api.openf1.org/v1/sessions?year=2026"

response=requests.get(url)
data=response.json()

df_quali=pd.DataFrame(data)
```

```python
df_quali.head(10)
```

```python
df_quali['date_end'] = pd.to_datetime(
    df_quali['date_end'],
    utc=True
)
```

```python
today = pd.Timestamp.now(tz='UTC')
df_quali=df_quali[df_quali['session_type']=='Qualifying']
df_quali=df_quali[df_quali['date_end']<today]
df_quali=df_quali[df_quali['is_cancelled']==False]
```

```python
df_quali.head(10)
```

```python
quali_final=[]
for _,session in df_quali.iterrows():
    session_key=session['session_key']
    url=f"https://api.openf1.org/v1/starting_grid?session_key={session_key}"
    response=requests.get(url)
    data=response.json()

    quali_final.extend(data)
df_starting_grid = pd.json_normalize(quali_final)
df_starting_grid=df_starting_grid.merge(
    df_quali[['session_key','circuit_key']],
    on='session_key',
    how='left'
)

```

```python
df_starting_grid.tail(10)
```

```python
df_avg_quali=df_starting_grid.groupby('driver_number').agg({'position':'mean'}).sort_values(by='position')
```

```python
df_avg_quali.head()
```

```python
combined_df['DriverNumber'] = (
    combined_df['DriverNumber']
    .astype(int)
)
```

```python
df_avg_quali=df_avg_quali.merge(
    combined_df[['DriverNumber','Abbreviation']],
    left_on='driver_number',
    right_on='DriverNumber',
    how='left'
)
```

```python
df_avg_quali=df_avg_quali.groupby('DriverNumber',as_index=False).agg({
        'position': 'mean',
        'Abbreviation': 'first'
    }).sort_values(by='position')
```

```python
url="https://api.openf1.org/v1/sessions?year=2026"

response=requests.get(url)
data=response.json()

df_race=pd.DataFrame(data)
```

```python
df_race.head()
```

```python
df_race['date_end'] = pd.to_datetime(
    df_race['date_end'],
    utc=True
)
```

```python
today = pd.Timestamp.now(tz='UTC')
df_race=df_race[df_race['session_name']=='Race']
df_race=df_race[df_race['date_end']<today]
df_race=df_race[df_race['is_cancelled']==False]
```

```python
df_race.head()
```

```python
df_delta=df_race.merge(
    df_starting_grid[['circuit_key','position']],
    on='circuit_key',
    how='left'
)
```

```python
df_delta.head()
```

```python
df_starting_grid.head()
```

```python
race_final = []

for _, session in df_race.iterrows():

    session_key = session["session_key"]

    url = (
        f"https://api.openf1.org/v1/"
        f"session_result?session_key={session_key}"
    )

    response = requests.get(url)

    data = response.json()

    race_final.extend(data)

df_race_res = pd.json_normalize(race_final)

df_race_res = df_race_res.merge(
    df_race[
        [
            "session_key",
            "circuit_key"
        ]
    ],
    on="session_key",
    how="left"
)
```

```python
df_race_res.head(20)
```

```python
df_delta = df_starting_grid.merge(
    df_race_res[
        [
            "meeting_key",
            "driver_number",
            "position",
            "points"
        ]
    ].rename(
        columns={
            "position": "race_position"
        }
    ),
    on=[
        "meeting_key",
        "driver_number"
    ],
    how="left"
)
```

```python
df_delta.head(20)
```

```python
max_grid = df_delta["position"].max()

df_delta["race_position"].fillna(max_grid+1,inplace=True)
```

```python
df_delta.isna().sum()
```

```python
df_delta["QualToRace"]=df_delta["position"]/df_delta["race_position"]
```
